```python
#Problem Statement 1:Smart Parking Lot Management System
def manage_parking(capacity, vehicle_logs):
    parked = 0
    peak_usage = 0
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        if parked > peak_usage:
            peak_usage = parked
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
    print("Peak Parking Usage:", peak_usage)
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)
```

```python
#Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    time = distance * 10
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    elif traffic == "Low":
        time += 5
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20
    print("Estimated Delivery Time:", time, "minutes")
distance = 8
traffic = "High"
weather = "Rainy"
estimate_delivery_time(distance, traffic, weather)
```

```python
#problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def analyze_seat_booking(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    print(f"Occupancy: {int(occupancy)}%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows")
    else:
        print("Show Status: Available"
```

```python
#Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    for cpu in cpu_readings:
        total += cpu
    average = total / len(cpu_readings)
    print(f"Average CPU Load: {int(average)}%")
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_resource_usage(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resource_usage(resource_usage)
```

```python
#Problem Statement 6: Online Event Registration Capacity Controller
def manage_event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # loop through each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
event_capacity = 100
total_registrations = 105

# function call (required)
manage_event_registration(event_capacity, total_registrations)
```
